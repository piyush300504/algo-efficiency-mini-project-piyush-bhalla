### Fibonacci (Dynamic Programming)

```python
def fibonacci_recursive(n):
    if n <= 1:
        return n
    return fibonacci_recursive(n - 1) + fibonacci_recursive(n - 2)
```

```python
def fibonacci_dp(n):
    if n <= 1:
        return n
    dp = [0] * (n + 1)
    dp[0], dp[1] = 0, 1
    for i in range(2, n + 1):
        dp[i] = dp[i - 1] + dp[i - 2]
    return dp[n]
```

```python
import time
import matplotlib.pyplot as plt

def fibonacci_recursive(n):
    if n <= 1:
        return n
    return fibonacci_recursive(n - 1) + fibonacci_recursive(n - 2)


def fibonacci_dp(n):
    if n <= 1:
        return n
    dp = [0] * (n + 1)
    dp[0], dp[1] = 0, 1
    for i in range(2, n + 1):
        dp[i] = dp[i - 1] + dp[i - 2]
    return dp[n]


def measure_time(func, n):
    start = time.time()
    func(n)
    end = time.time()
    return end - start


dp_ns = list(range(1000, 10001, 1000)) 

recursive_times = [measure_time(fibonacci_recursive, n) for n in recursive_ns]
dp_times = [measure_time(fibonacci_dp, n) for n in dp_ns]


plt.figure(figsize=(10, 6))

plt.plot(recursive_ns, recursive_times, marker="o", color="red", label="Recursive Fibonacci")

plt.plot(dp_ns, dp_times, marker="o", color="blue", label="DP Fibonacci")

plt.xlabel("Input Size (n)")
plt.ylabel("Execution Time (seconds)")
plt.title("Execution Time vs Input Size for Fibonacci (Recursive vs DP)")
plt.legend()
plt.grid(True)
plt.show()
```

### Merge Sort

```python
def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        left = arr[:mid]
        right = arr[mid:]

        merge_sort(left)
        merge_sort(right)

        i = j = k = 0
        while i < len(left) and j < len(right):
            if left[i] < right[j]:
                arr[k] = left[i]
                i += 1
            else:
                arr[k] = right[j]
                j += 1
            k += 1

        while i < len(left):
            arr[k] = left[i]
            i += 1
            k += 1
        while j < len(right):
            arr[k] = right[j]
            j += 1
            k += 1
    return arr
```

### Quick Sort

```python
def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quick_sort(left) + middle + quick_sort(right)
```

### Insertion Sort

```python
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr
```

### Bubble Sort

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr
```

### Selection Sort

```python
def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr
```

```python
import time
import random
import matplotlib.pyplot as plt

def measure_time(func, arr):
    start = time.time()
    func(arr.copy()) 
    end = time.time()
    return end - start

input_sizes = [100, 500, 1000, 2000]

merge_times = []
quick_times = []
insertion_times = []
bubble_times = []
selection_times = []

for n in input_sizes:
    arr = [random.randint(1, 10000) for _ in range(n)]
    merge_times.append(measure_time(merge_sort, arr))
    quick_times.append(measure_time(quick_sort, arr))
    insertion_times.append(measure_time(insertion_sort, arr))
    bubble_times.append(measure_time(bubble_sort, arr))
    selection_times.append(measure_time(selection_sort, arr))


plt.figure(figsize=(10, 6))
plt.plot(input_sizes, merge_times, marker="o", label="Merge Sort")
plt.plot(input_sizes, quick_times, marker="o", label="Quick Sort")
plt.plot(input_sizes, insertion_times, marker="o", label="Insertion Sort")
plt.plot(input_sizes, bubble_times, marker="o", label="Bubble Sort")
plt.plot(input_sizes, selection_times, marker="o", label="Selection Sort")

plt.xlabel("Input Size (n)")
plt.ylabel("Execution Time (seconds)")
plt.title("Execution Time vs Input Size for Sorting Algorithms")
plt.legend()
plt.grid(True)
plt.show()
```

### Binary Search

```python
def binary_search(arr, target):
    low, high = 0, len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1
```

```python
import time
import matplotlib.pyplot as plt
import random

def measure_time(func, arr, target):
    start = time.time()
    func(arr, target)
    end = time.time()
    return end - start

input_sizes = [1000, 5000, 10000, 50000, 100000, 500000, 1000000]

times = []

for n in input_sizes:
    arr = sorted(random.sample(range(1, n * 2), n))  # sorted array
    target = random.choice(arr)  # search for random element
    times.append(measure_time(binary_search, arr, target))

plt.figure(figsize=(10, 6))
plt.plot(input_sizes, times, marker="o", color="blue", label="Binary Search")

plt.xlabel("Input Size (n)")
plt.ylabel("Execution Time (seconds)")
plt.title("Execution Time vs Input Size for Binary Search")
plt.legend()
plt.grid(True)
plt.show()
```
